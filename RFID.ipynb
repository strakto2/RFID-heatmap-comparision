```python
#################################################################################################################
########################################### Antena 1 example ####################################################
#################################################################################################################


img = cv2.imread('Fig3_resized.jpeg')


grid_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20,8))
plt.imshow(grid_RGB) # Printing the original picture


grid_HSV = cv2.cvtColor(grid_RGB, cv2.COLOR_RGB2HSV) # Converting to HSV
lower1_red = np.array([0, 40, 40])
upper1_red = np.array([10, 255, 255])


lower2_red = np.array([160,40,40])
upper2_red = np.array([179,255,255])

lower_mask = cv2.inRange(grid_HSV, lower1_red, upper1_red)
upper_mask = cv2.inRange(grid_HSV, lower2_red, upper2_red)

mask=lower_mask + upper_mask;

res = cv2.bitwise_and(img, img, mask=mask)

print("Red Part of Image (-1 to -20 dBm)")
plt.figure(figsize=(20,8))
plt.imshow(res)

RED_perc = (mask>0).mean()
print(RED_perc)
```

```python
grid_HSV = cv2.cvtColor(grid_RGB, cv2.COLOR_RGB2HSV) # Converting to HSV
lower_yellow = np.array([11, 93, 0])
upper_yellow = np.array([25, 255, 255])

mask= cv2.inRange(grid_HSV, lower_yellow, upper_yellow)
res = cv2.bitwise_and(img, img, mask=mask)

print("Orange Part of Image (-20 to -30 dBm)")
plt.figure(figsize=(20,8))
plt.imshow(res)

yellow_perc = (mask>0).mean()
print(yellow_perc)
```
